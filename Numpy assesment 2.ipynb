Write the applications of numpy.
1. Scientific Computing
- Provides efficient handling of large datasets and numerical operations.
- Used in physics, chemistry, biology, and engineering simulations.
- Example: Solving differential equations, matrix operations, Fourier transforms.

2. Data Science & Machine Learning
- Forms the foundation of libraries like Pandas, scikit-learn, TensorFlow, PyTorch.
- Used for preprocessing data: normalization, vectorization, feature scaling.
- Example: Converting raw datasets into structured arrays for ML models.

3. Image Processing
- Images can be represented as NumPy arrays (pixels as numerical values).
- Enables operations like filtering, transformations, and compression.
- Example: Grayscale conversion, edge detection, resizing.

```python
import numpy as np
```

```python
data=np.genfromtxt("C:\\Users\\srava_kv\\customer details\\customer_purchases.csv",delimiter=",",encoding=None,dtype=None,names=True)
```

```python
data
```

```python
unique_data=np.unique(data["CustomerID"])
```

```python
print(unique_data)
```

```python
total_amount_spend_customer={}
```

```python
for customer in unique_data:
    total_amount_spend_customer[customer]=np.sum(data[data["CustomerID"]==customer]["AmountSpent"])
```

```python
total_amount_spend_customer
```

```python
unique_dates=np.unique(data["Date"])
```

```python
unique_dates
```

```python
daily_sales=np.zeros(unique_dates.size)
```

```python
daily_sales
```

```python
for i,date in enumerate(unique_dates):
    daily_sales[i]=np.sum(data[data["Date"]==date]["AmountSpent"])
```

```python
daily_sales
```

```python
highest_sale=np.argmax(daily_sales)
```

```python
print(highest_sale)
```

```python
highest_sales_day=unique_dates[highest_sale]
```

```python
print(highest_sales_day)
```

```python
import csv

with open("Purchase_Analysis.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["CustomerID", "Total_Amount_Spent"])
    for id, total in total_amount_spend_customer.items():
        writer.writerow([id, total])
    writer.writerow([])  # Blank row
    writer.writerow(["Date", "Total_Sales"])
    for date, sales in zip(unique_dates, daily_sales):
        writer.writerow([date, sales])
```

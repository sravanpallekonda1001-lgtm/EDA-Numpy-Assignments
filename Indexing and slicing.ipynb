## Part A: Integer Indexing & Slicing

1. Integer Indexing & Negative Indexing
Input
- A = np.array([100, 200, 300, 400, 500])
Expected Output
- First two elements → [100, 200]
- Last three elements → [300, 400, 500]
- Middle element → 300

```python
import numpy as np
```

```python
arr=np.array([100,200,300,400,500])
```

```python
print("first two elements -->",arr[:2])
print("last three elements -->",arr[-3:])
print("middle element -->",arr[len(arr)//2])
```

## Part B: Boolean Indexing

5. Condition-Based Filtering
Input
- A = np.array([5, 12, 17, 3, 8, 20])
- Expected Output
- Values > 10 → [12, 17, 20]
- Values ≤ 8 → [5, 3, 8]

```python
A = np.array([5, 12, 17, 3, 8, 20])
print(A[A>10])
print(A[A<=8])
```

6. Multiple Boolean Conditions
- Input

- Expected Output
 [20, 25, 30, 35]

```python
A = np.array([10, 15, 20, 25, 30, 35])
print(A[A>15])
```

7. Boolean Filtering + Counting
- Input
- A = np.array([3, -2, 0, 5, -7, 8])
- Expected Output
- Non-negative values → [3, 0, 5, 8]
Negative count → 2

```python
A = np.array([3, -2, 0, 5, -7, 8])
count=np.sum(A[A<0])
print("Non negatice values -->",A[A>=0],"Negativve count -->",count)
```

8. Boolean Replacement
- Input
- A = np.array([1, 2, 3, 4, 5, 6, 7])
- Expected Output
 [-1, 2, -1, 4, -1, 6, -1]

```python
A = np.array([1, 2, 3, 4, 5, 6, 7])
result=np.where(A%2!=0,-1,A)
print(result)
```

## Part C: Fancy Indexing


9. Index List Selection
- Input
- A = np.array([11, 22, 33, 44, 55, 66])
- Expected Output
 [22, 44, 66]


```python
A = np.array([11, 22, 33, 44, 55, 66])
result=A[np.where(A%2==0)]
print(result)
```

## Part D: Boolean Indexing vs np.where()

13. Filtering vs Transforming
- Input
 A = np.array([4, 9, 2, 7, 12, 3])
- Expected Output
Boolean indexing → [9, 7, 12]
- np.where() → [0, 1, 0, 1, 1, 0]

```python
A = np.array([4, 9, 2, 7, 12, 3])
result=A[A>5]
print(result)
result2=np.where(A>5,1,0)
print(result2)
```

14. Replace Negatives (Two Methods)
- Input
 A = np.array([-5, 10, -3, 8, -1])
- Expected Output
 [0, 10, 0, 8, 0]

```python
A = np.array([-5, 10, -3, 8, -1])
result=np.where(A<0,0,A)
print(result)
```

15. Labeling Data with np.where()
- Input
 scores = np.array([45, 60, 30, 85, 90])
- Expected Output
- Filtered scores → [60, 85, 90]
- Labels → ["Fail", "Pass", "Fail", "Pass", "Pass"]

```python
scores = np.array([45, 60, 30, 85, 90])
filter_score=scores[scores>55]
print("Filter score:",filter_score)
result=[]
for i in scores:
    if i<50:
        result.append('''Fail''')
    else:
        result.append("Pass")
print("Labels",result)
```
